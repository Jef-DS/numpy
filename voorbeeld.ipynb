Je kunt hiervoor de functie *argmax()* gebruiken. Die hebben we niet gezien, maar het zal wel duidelijk zijn wat die doet (denk aan argsort())

datums = np.genfromtxt(BESTAND, delimiter=';', skip_header=1, dtype=np.datetime64, usecols=0)

# Oplossing 4
Let op met de dataset: De namen bevatten een komma. Dat is geen probleem voor de CSvreader van Python (en ook niet voor Pandas) aangezien de teksten tussen dubbele aanhalingstekens staan. Maar *genfromtext()* snapt dit niet. Mijn oplossing is dat we het bestand lezen met Python en bewaren als CSV-bestand met als delimiter een puntkomma.

Met np.unique() kunnen we de unieke waarden opvragen en (met return_counts=True) de aantallen
```
from csv import DictReader, DictWriter
# Schrijf hier de code.
BESTAND = "Titanic-Dataset.csv"
TITANIC_CSV = "Titanic_puntkomma.csv"
# controleer de inhoud
#with open(BESTAND, 'r') as f:
#    for _ in range(5):
#        print(f.readline(), end="")
with open(BESTAND, 'r', newline="") as f_in:
    reader = DictReader(f_in, delimiter=',')
    with open (TITANIC_CSV, mode='w', newline="") as f_out:
        writer = DictWriter(f_out, delimiter=';', fieldnames=reader.fieldnames)
        writer.writeheader()
        for regel in reader:
            writer.writerow(regel)
geslacht = np.genfromtxt(TITANIC_CSV, skip_header=1, dtype=str, usecols=4, delimiter=';')
geslachten, aantallen = np.unique(geslacht, return_counts=True)
aantal_mannen = aantallen[geslachten=='male'][0]
aantal_vrouwen = aantallen[geslachten=='female'][0]
```
# Oplossing 5
Er zijn hier verschillende mogelijke manieren om dit op te lossen. Maar dit is een voorbeeld. Ik maak hier gebruik van het feit dat de kolom 'overleefd' de getallen 1 of 0 bevat. Vandaar dat we *np.count_nonzero()* kunnen gebruiken.
````
TITANIC_CSV = "Titanic_puntkomma.csv"
arr_geslacht = np.genfromtxt(TITANIC_CSV, skip_header=1, dtype=str, usecols=4, delimiter=';')
arr_overleefd = np.genfromtxt(TITANIC_CSV, skip_header=1, usecols=1, delimiter=';')
arr_klasse = np.genfromtxt(TITANIC_CSV, skip_header=1, usecols=2, dtype=np.int8, delimiter=';')
totaal_mannen_per_klasse = {}
totaal_vrouwen_per_klasse = {}
totaal_mannen_overleefd_per_klasse = {}
totaal_vrouwen_overleefd_per_klasse = {}
for klasse in (1, 2, 3):
    totaal_mannen_per_klasse[klasse] = np.count_nonzero((arr_geslacht=='male') & (arr_klasse==klasse))
    totaal_vrouwen_per_klasse[klasse] = np.count_nonzero((arr_geslacht=='female') & (arr_klasse== klasse))
    totaal_mannen_overleefd_per_klasse[klasse] = np.count_nonzero(arr_overleefd[(arr_geslacht=='male') & (arr_klasse==klasse)])
    totaal_vrouwen_overleefd_per_klasse[klasse] = np.count_nonzero(arr_overleefd[(arr_geslacht=='female') & (arr_klasse==klasse)])
percentage_vrouwen_overleefd_per_klasse = {}
percentage_mannen_overleefd_per_klasse = {}
for klasse in (1, 2, 3):
    percentage_mannen_overleefd_per_klasse[klasse] = totaal_mannen_overleefd_per_klasse[klasse] / totaal_mannen_per_klasse[klasse] * 100
    percentage_vrouwen_overleefd_per_klasse[klasse] = totaal_vrouwen_overleefd_per_klasse[klasse] / totaal_vrouwen_per_klasse[klasse] * 100
