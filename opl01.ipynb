# Vraag 5

```python
import numpy as np
dagen = np.array(["Maandag", "Dinsdag", "Woensdag", "Donderdag", "Vrijdag", "Zaterdag", "Zondag"])
steden = np.array(["Antwerpen", "Leuven", "Mechelen"])
rng = np.random.default_rng(42)
temperaturen = rng.integers(-5, 10, (steden.size, dagen.size))
#print('De temperaturen:', temperaturen, sep='\n')

#Pas de functie aan zodat ze een array met de gemiddelde temperaturen voor elke dag teruggeeft
def gemiddelde_temperatuur_per_dag(temps):
    return np.mean(temps, axis=0)

for dag in dagen:
    print(f"{dag:>10s}", end='')
print()
for temp in gemiddelde_temperatuur_per_dag(temperaturen):
    print(f"{temp:10.1f}", end='')
```

```python
import numpy as np
dagen = np.array(["Maandag", "Dinsdag", "Woensdag", "Donderdag", "Vrijdag", "Zaterdag", "Zondag"])
steden = np.array(["Antwerpen", "Leuven", "Mechelen"])
rng = np.random.default_rng(42)
temperaturen = rng.integers(-5, 10, (steden.size, dagen.size))
#print('De temperaturen:', temperaturen, sep='\n')

def hoogste_temperatuur_op_weekdag(temp):
    return np.max(temp[:, :5], axis=1)

for stad, temp in zip(steden, hoogste_temperatuur_op_weekdag(temperaturen)):
    print(f"{stad}: {temp}°C")
```

# Vraag 9

```python
import numpy as np
import warnings
dagen = np.array(["Maandag", "Dinsdag", "Woensdag", "Donderdag", "Vrijdag", "Zaterdag", "Zondag"])
steden = np.array(["Antwerpen", "Leuven", "Mechelen"])
rng = np.random.default_rng(42)
temperaturen = rng.integers(-5, 10, (steden.size, dagen.size))
#print('De temperaturen:', temperaturen, sep='\n')

#Pas de volgende functie aan
def gemiddelde_temperatuur_in(arr_temps, arr_steden, locatie):
    is_stad = arr_steden == locatie
    if np.count_nonzero(is_stad) == 0:
        raise IndexError()
    return np.mean(arr_temps[is_stad])

with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    for stad in ('Leuven', 'Mechelen', 'Brussel'):
        try:
            gemiddelde_temperatuur = gemiddelde_temperatuur_in(temperaturen, steden, stad)
            print(f'De gemiddelde temperatuur in {stad} was {gemiddelde_temperatuur:.1f}°C')
        except IndexError:
            print(f'Voor {stad} hebben we geen temperatuur.')
```

# Vraag 10

```python
from csv import DictReader, DictWriter
TITANIC_CSV = 'Titanic-Dataset.csv'
TITANIC_PSV = 'Titanic.csv'  #puntkomma separated values
with open(TITANIC_CSV, newline='') as f_in:
    reader = DictReader(f_in, delimiter=',')
    with open(TITANIC_PSV, mode='w') as f_out:
        writer = DictWriter(f_out, delimiter=';', fieldnames=reader.fieldnames)
        writer.writeheader()
        for regel in reader:
            writer.writerow(regel)

namen = np.genfromtxt(TITANIC_PSV, delimiter=';', skip_header=1, usecols=3, dtype=str)
prijzen = np.genfromtxt(TITANIC_PSV, delimiter=';', skip_header=1, usecols=-3)
naam = 'Cameron'
is_cameron = np.strings.startswith(namen, naam)
volledige_naam = namen[is_cameron][0]
prijs = prijzen[is_cameron][0]
print(f'Passagier "{volledige_naam}" betaalde £{prijs}.')
```
