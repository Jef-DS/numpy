```
Op zaterdag 2026-02-07 is er gedurende de hele week voor €12487 verkocht.
Op zaterdag 2026-02-14 is er gedurende de hele week voor €12586 verkocht.
Op zaterdag 2026-02-21 is er gedurende de hele week voor €12433 verkocht.
Op zaterdag 2026-02-28 is er gedurende de hele week voor €13517 verkocht.
```

````python
from datetime import date, timedelta
verkopen = [0, 2547, 2309, 1877, 1866, 2717, 1171, 0, 1402, 1188, 2052,
       2951, 2471, 2522, 0, 2572, 2026, 1256, 2679, 1900, 2000, 0,
       1365, 2853, 2563, 2287, 1804, 2645]
dagen = [date(2026, 2, 1) + timedelta(dag) for dag in range(28)]
MAANDOMZET_CSV = 'maandomzet.csv'
with open(MAANDOMZET_CSV, mode='w') as f:
    print('dag;verkoopcijfer', file=f)
    for dag, cijfer in zip(dagen, verkopen):
        print(f"{dag};{cijfer}", file=f)
````
