## Vraag 2
Je krijgt van mij drie arrays met de voornamen, de achternamen en de geboortedatums van de drie beste zangeressen van de lage landen. (je wist dat je dit soort vraag zou krijgen). De bedoeling is dat je de variabelen *voornamen_sorted*, *achternamen_sorted* en *geboortedatums_sorted* correct invult zodat de zangeressen gesorteerd op achternaam worden getoond. De output moet er dus als volgt uitzien:
```
Kathleen Aerts is geboren op 18-06-1978
Karen Damen is geboren op 28-10-1974
Kristel Verbeke is geboren op 10-12-1975
```



````python
import numpy as np
from datetime import date
voornamen = np.array(['Karen', 'Kristel', 'Kathleen'])
achternamen = np.array(['Damen', 'Verbeke', 'Aerts'])
geboortedatums = np.array(['1974-10-28', '1975-12-10', '1978-06-18'], dtype=np.datetime64)
#Hoe moet je de volgende 3 statements wijzigen om te sorteren op achternaam
voornamen_sorted = voornamen
achternamen_sorted = achternamen
geboortedatums_sorted = geboortedatums
for voornaam, achternaam, geboortedatum in zip(voornamen_sorted, achternamen_sorted, geboortedatums_sorted):
  print(f'{voornaam} {achternaam} is geboren op {geboortedatum.astype(date).strftime("%d-%m-%Y")}')
````
