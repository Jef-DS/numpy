# Vragen examenresultaten
## Vooraf
Gebruik de volgende code om een bestand met examenresultaten te maken.

````python
from csv import DictWriter
bestand = 'examen.csv'
data = [
    {'naam':'Jan', 'Python':13, 'NumPy':'12', 'Pandas': 12},
    {'naam':'Piet', 'Python':17, 'NumPy':15, 'Pandas': 13},
    {'naam':'Joris', 'Python':7, 'NumPy':9, 'Pandas': '10'},
    {'naam':'Corneel', 'Python': 14, 'NumPy': 10, 'Pandas': 19},
        ]
with open(bestand, 'w', newline='', encoding='utf8') as f:
  writer = DictWriter(f, fieldnames=data[0].keys(), delimiter=';')
  writer.writeheader()
  writer.writerows(data)
data = None
with open(bestand, 'r', encoding='utf8') as f:
  for regel in f:
    print(regel, end='')
````
