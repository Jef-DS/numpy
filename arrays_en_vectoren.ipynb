# Vectoren
## Een vector is een pijl
Een vector wordt bepaald door 2 eigenschappen:
1. een lengte
1. een richting

Vandaar dat we een vector kunnen tekenen als een pijl. Standaard tekenen we die pijl in de oorsprong(0,0), maar eigenlijk kan die overal getekend worden. In de volgende grafiek wordt dezelfde vector drie keer getekend, telkens vertrekkend in een andere oorsprong. Belangrijk: dit is drie keer dezelfde vector.

In NumPy is een vector een 1-dimensionele array.

```python
import matplotlib.pyplot as plt
import numpy as np
vector = np.array([3, 4])
plt.quiver(0, 0, *vector, angles='xy', scale_units='xy', scale=1)
plt.quiver(2, 4, *vector, angles='xy', scale_units='xy', scale=1)
plt.quiver(3, 1, *vector, angles='xy', scale_units='xy', scale=1)
plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.grid()
plt.show()
```

## De lengte van een vector
De lengte van de pijl komt overeen met de lengte van de vector. We kunnen die zelf berekenen met behulp van de stelling van Pythagoras: $z = \sqrt{x^2 + y^2}$

```python
import numpy as np
vector = np.array([3, 4])
lengte = np.sqrt(np.sum(vector**2))
print('Volgens Pythagoras:')
print(f'{lengte=}')
lengte2 = np.linalg.norm(vector)
print('Met np.linalg.norm')
print(f'{lengte2=}')
```

## Vector vermenigvuldigen met een getal
Wanneer we een vector vermenigvuldigen met een positief getal, veranderen we de lengte (maar niet de richting). Vermenigvuldigen met een negatief getal keert de richting om.

```python
import matplotlib.pyplot as plt
import numpy as np
vector = np.array([1, 1])
vector_2 = vector*2
vector_neg = vector*-1
print(f'{vector=}')
print(f'{vector_2=}')
print(f'{np.linalg.norm(vector)=}')
print(f'{np.linalg.norm(vector_2)=}')
print(f'{np.linalg.norm(vector_neg)=}')
plt.quiver(0, 0, *vector, angles='xy', scale_units='xy', scale=1, color='red', label='oorspronkelijk')
plt.quiver(0, 0, *vector_2, angles='xy', scale_units='xy', scale=1, color='green', label = ' x 2')
plt.quiver(0, 0, *vector_neg, angles='xy', scale_units='xy', scale=1, color='blue', label= ' x -1')
plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.legend()
plt.grid()
plt.show()
```

## Vectoren optellen
In een grafiek kunnen we vectoren optellen door ze 'achter elkaar' te zetten. We laten de tweede vector beginnen in het eindpunt van de eerste.

```python
import matplotlib.pyplot as plt
import numpy as np
vector1 = np.array([1, 2])
vector2 = np.array([2, 2])
vector3 = vector1 + vector2
print(f'{vector1=}')
print(f'{vector2=}')
print(f'{vector3=}')

plt.quiver(0, 0, *vector1, angles='xy', scale_units='xy', scale=1, color='red', label='vector1')
plt.quiver(*vector1, *vector2, angles='xy', scale_units='xy', scale=1, color='green', label = 'vector2')
plt.quiver(0, 0, *vector3, angles='xy', scale_units='xy', scale=1, color='blue', label= 'vector1 + vector2')
plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xlim(-0.5, 6)
plt.ylim(-0.5, 6)
plt.legend()
plt.grid()
plt.show()
```

## Herhaling: een stelsel van vergelijkingen (matrices)
We hebben het volgende stelsel al gezien.
$$x_1 + x_2 = 3 \\
3x_1 -2x_2 = 4$$
We kunnen dit voorstellen in 'matrix-vorm':
$$\begin{bmatrix}1 & 1 \\3 & -2 \end{bmatrix} \begin{bmatrix}x_1 \\ x_2\end{bmatrix} = \begin{bmatrix}3 \\ 4\end{bmatrix}$$

We zien hier twee vectoren ($x_1$, $x_2$) en (3, 4). De tweedimensionele array noemen we een matrix.

We kunnen dit stelsel ook schrijven als:
$$ Ax = y$$
We hebben gezien dat we dit kunnen oplossen door de inverse van de matrix te nemen:
$$ x = A^{-1} y$$
Maar wat wil dat zeggen wanneer we dit in een grafiek bekijken?


```python
import matplotlib.pyplot as plt
import numpy as np
y_vect = np.array([3, 4])
A_mat = np.array([[1, 1], [3, 2]])
A_inv = np.linalg.inv(A_mat)
x_vect = A_inv @ y_vect
print(f'{x_vect=}')
plt.quiver(0, 0, *y_vect, angles='xy', scale_units='xy', scale=1, color='red', label='y')
plt.quiver(0, 0, *x_vect, angles='xy', scale_units='xy', scale=1, color='blue', label= 'x')
plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xlim(-2.5, 5)
plt.ylim(-0.5, 5)
plt.legend()
plt.show()
```

## Een matrix 'transformeert' een vector
Door de y-vector te vermenigvuldigen met de inverse van A krijgen we de x-vector. Door de x-vector te vermenigvuldigen met A krijgen de y-vector. De vraag die we dus proberen op te lossen wanneer we een stelsel van vergelijkingen in NumPy oplossen is dus: wat is de vector (x-vector) die vermenigvuldigd met de matrix A de y-vector geeft.

Wanneer we een vector vermenigvuldigen met een matrix krijgen we een tweede vector. Wanneer we die tweede vector vermenigvuldigen met de inverse van de matrix, krijgen we de eerste vector terug.

Let op: we moeten in NumPy een matrixvermenigvuldiging uitvoeren: @. Die wordt als volgt gedefinieerd:
$$\begin{bmatrix}1 & 1 \\3 & -2 \end{bmatrix} @ \begin{bmatrix}x_1 \\ x_2\end{bmatrix} = \begin{bmatrix}1x_1 + 1x_2 \\3x_1 + -2x_2 \end{bmatrix}$$

```python
import numpy as np
y_vect = np.array([3, 4])
A_mat = np.array([[1, 1], [3, 2]])
A_inv = np.linalg.inv(A_mat)
x_vect = A_inv @ y_vect
print(f'{x_vect=}')
y_vect = A_mat @ x_vect
print(f'{y_vect=}')
```

## Enkele speciale transformatiematrices: eenheidsmatrix en spiegeling
Er zijn ook enkele speciale matrices. Net zoals we voor de optelling een getal hebben dat 'niets doet', namelijk 0 (en voor de vermenigvuldiging 1) bestaat dat ook voor de matrixvermenigvuldiging: de eenheidsmatrix. Dat is een matrix met overal 0, behalve op de diagonaal. Daar staat een 1. In NumPy kunnen we die maken met *np.eye()*. Als argument geven we mee hoeveel rijen (of kolommen) we willen.

In de volgende code is er nog een tweede transformatiematrix gedefinieerd: spiegeling rond de diagonaal in het assenkruis.

```python
import matplotlib.pyplot as plt
import numpy as np
vect = np.array([1, 2])
eenheidsmatrix = np.eye(2) #np.array([[1, 0], [0, 1]])
print(f'{vect=}')
print(f'eenheidsmatrix:\n {eenheidsmatrix}')
print(f'{vect @ eenheidsmatrix=}')
spiegelmatrix = np.array([[0, 1], [1, 0]]) #spiegelt rond de 45 lijn
print(f'spiegelmatrix:\n {spiegelmatrix}')
vect_gespiegeld = vect @ spiegelmatrix
print(f'{vect @ spiegelmatrix=}')
plt.quiver(0, 0, *vect, angles='xy', scale_units='xy', scale=1, color='red', label='oorspronkelijk')
plt.quiver(0, 0, *vect_gespiegeld, angles='xy', scale_units='xy', scale=1, color='blue', label='gespiegeld')
plt.gca().set_aspect('equal')
plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xlim(-0.5, 5)
plt.ylim(-0.5, 5)
plt.legend()
plt.show()
```

## Een rotatiematrix
Een matrix die over een hoek $\theta$ in tegenwijzerzin roteert, ziet er als volgt uit:
$$\begin{bmatrix}cos\theta & -sin\theta \\sin\theta & cos\theta \end{bmatrix}$$

Let op: de functies np.cos() en np.sin() verwachten de hoek in radialen en niet in graden: radialen = graden/180 * $\pi$

```python
import matplotlib.pyplot as plt
import numpy as np
vect = np.array([2, 1])
graden = 60
hoek = graden/180 * np.pi
rotatie = np.array([[np.cos(hoek), -np.sin(hoek)], [np.sin(hoek), np.cos(hoek)]])
nieuwe_vect = rotatie @ vect
print(f'{vect=}')
print(f'{rotatie=}')
print(f'{nieuwe_vect=}')
plt.quiver(0, 0, *vect, angles='xy', scale_units='xy', scale=1, color='red', label='oorspronkelijk')
plt.quiver(0, 0, *nieuwe_vect, angles='xy', scale_units='xy', scale=1, color='blue', label=f'gedraaid over {graden}')
plt.gca().set_aspect('equal')
plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xlim(-0.5, 3)
plt.ylim(-0.5, 3)
plt.legend()
plt.show()
```

##De determinant van een transformatiematrix
We kunnen twee vectoren tegelijk transformeren met een matrix door ze samen te voegen. Met twee vectoren kunnen we een parallellogram vormen. De determinant van een matrix bepaalt de schaalfactor waarmee het oppervlak van dat parallellogram geschaald wordt.

We kunnen dat het duidelijkste uitleggen door twee vectoren te nemen die een vierkant vormen: [1, 0] en [0, 1]

```python
import matplotlib.pyplot as plt
import numpy as np
x_vect = np.array([1, 0])
y_vect = np.array([0, 1])
oppervlakte1 = np.linalg.norm(x_vect) * np.linalg.norm(y_vect)
print(f'oorspronkelijke oppervlakte: {oppervlakte1}')
vectoren = np.column_stack([x_vect, y_vect])
vectoren = np.column_stack([x_vect, y_vect])
matrix = np.eye(2) * 2
geschaalde_vectoren = matrix @ vectoren
x_geschaald = geschaalde_vectoren[:, 0]
y_geschaald = geschaalde_vectoren[:, 1]
oppervlakte2 = np.linalg.norm(x_geschaald) * np.linalg.norm(y_geschaald)
print(f'geschaalde oppervlakte: {oppervlakte2}')
print(f'{np.linalg.det(matrix)=}')
plt.quiver(0, 0, *x_vect, angles='xy', scale_units='xy', scale=1, color='red', label='x')
plt.plot([0, 1], [1, 1], color='red', linestyle='-.')
plt.plot([1, 1], [0, 1], color='red', linestyle='-.')
plt.quiver(0, 0, *y_vect, angles='xy', scale_units='xy', scale=1, color='red', label='y')
plt.quiver(0, 0, *x_geschaald, angles='xy', scale_units='xy', scale=1, color='blue', label='x_geschaald')
plt.quiver(0, 0, *y_geschaald, angles='xy', scale_units='xy', scale=1, color='blue', label='y_geschaald')
plt.plot([0, 2], [2, 2], color='blue', linestyle='-.')
plt.plot([2, 2], [0, 2], color='blue', linestyle='-.')
plt.gca().set_aspect('equal')
plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xlim(-0.5, 3)
plt.ylim(-0.5, 3)
plt.legend()
plt.show()
```

## Niet-inverteerbare (singuliere) matrix
Niet elke matrix heeft een inverse. Wanneer de determinant van een matrix 0 is, kan de matrix niet genverteerd worden. Een matrix die niet genverteerd kan worden, noemen we ook een singuliere matrix. Een voorbeeld:
$$\begin{bmatrix}1 & 0 \\ 0 & 0 \end{bmatrix}$$

Wanneer we het resultaat van de transformatie bekijken van de twee vectoren, zien we dat ze allebei als resultaat [1, 0] geven. (de rode pijl(1,0) is niet zichtbaar omdat de blauwe pijl erover getekend is).

Bedenk nu wat de inverse transformatie zou moeten doen: de vector [1, 0] omzetten naar [1, 2], maar diezelfde vector moet met de inverse transformatie ook [1, 3] opleveren. Dat kan niet. Een transformatie van een vector kan geen twee resultaten opleveren. Vandaar dat de inverse van de matrix niet bestaat.

Wanneer we kijken naar de oppervlakte die gevormd wordt door de twee getransformeerde vectoren ([1, 0] en [1, 0]), zien we dat die oppervlakte 0 is. Twee vectoren die in elkaars verlengde liggen vormen een parallellogram met oppervlakte 0. De schaalfactor (determinant) van de transformatiematrix is 0. En vandaar dat een matrix met schaalfactor 0 geen inverse heeft.

```python
import matplotlib.pyplot as plt
import numpy as np
vect_1 = np.array([1, 2])
vect_2 = np.array([1, 3])
vect = np.column_stack([vect_1, vect_2])
matrix = np.array([[1, 0], [0, 0]])
print(f'{np.linalg.det(matrix)=}')
vect_trans = matrix @ vect
vect_1_trans = vect_trans[:, 0]
vect_2_trans = vect_trans[:, 1]
print(f'{vect_1_trans=}')
print(f'{vect_2_trans=}')
plt.quiver(0, 0, *vect_1, angles='xy', scale_units='xy', scale=1, color='red', label='x')
plt.quiver(0, 0, *vect_2, angles='xy', scale_units='xy', scale=1, color='blue', label='y')
plt.quiver(0, 0, *vect_1_trans, angles='xy', scale_units='xy', scale=1, color='red', label='x transformed')
plt.quiver(0, 0, *vect_2_trans, angles='xy', scale_units='xy', scale=1, color='blue', label='y transformed')
plt.gca().set_aspect('equal')
plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.xlim(-0.5, 3)
plt.ylim(-0.5, 3)
plt.legend()
plt.show()
```
