##Oef3
In de volgende code verander ik de array zodat er een rand met 0-waarden wordt gemaakt. Wanneer je de code bekijkt, zul je merken dat de vier hoekpunten ([0,0], [0, -1], [-1, 0] en [-1, -1]) twee keer de waarde 0 krijgen.

Je kunt dat zien door `arr[rijen, :] = 0` in commentaar te zetten (de hoekpunten zijn 0) of door `arr[:, kolommen] = 0` in commentaar te zetten (dan zijn de hoekpunten nog steeds 0)

Dat is dubbel werk maar het zal geen invloed hebben op de performantie. Hoe kan je echter de code aanpassen zodat de hoekpunten maar éénmaal de waarde 0 krijgen en er dus geen dubbel werk gebeurt.
1. beide statements vervangen door: arr[rijen, kolommen] = 0
1. arr[rijen, rijen] = 0 en arr[kolommen, kolommen] = 0
1. arr[rijen, 1:-1] = 0 en arr[1, -1, kolommen] = 0
1. arr[rijen, :] = 0 en arr[1:-1, kolommen] = 0


````python
import numpy as np
rng = np.random.default_rng(42)
arr = rng.integers(0, 100, size=(5, 5))
print('arr:\n', arr)
rijen = np.array([0, -1])
print(f'{rijen.shape=}')
kolommen = np.array([0, -1])
print(f'{kolommen.shape=}')
arr[rijen, :] = 0
arr[:, kolommen] = 0
print('arr:\n', arr)
````

##Gewoon ter illustratie (geen vraag)
Wanneer we van een kleurenfoto een foto met grijswaarden willen maken, kunnen we de RGB-waarden aan elkaar gelijk stellen door ze te delen door 3.

Maar we kunnen ook rekening houden met de gevoeligheid van het menselijk ook voor kleuren. We zijn bijvoorbeeld meer gevoelig voor de kleur groen, dus vandaar dat die meer meetelt. (zie bijvoorbeeld [How to Convert an RGB Image to Grayscale](https://brandonrohrer.com/convert_rgb_to_grayscale.html))

````python
#Gray = 0.212 * Red + 0.715 * Green + 0.072 * Blue
import matplotlib.pyplot as plt
import numpy as np
import PIL.Image as Image
foto = Image.open(K3_JPG)
arr = np.array(foto)
luminosity = np.array([0.212, 0.715, 0.072])
average = np.array([1/3, 1/3, 1/3])
arr_lum = arr @ luminosity #of np.dot(arr, luminosity)
arr_av = arr @ average
print(arr_lum[0,0])
print(arr_av[0,0])
fig, axes = plt.subplots(1, 2, figsize=(15, 7)) #figsize: totale grootte figuur in inches(2,54cm)
axes[0].imshow(arr_lum, cmap='gray')
axes[1].imshow(arr_av, cmap='gray')
plt.show()
````
